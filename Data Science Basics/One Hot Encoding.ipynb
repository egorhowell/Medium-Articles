# One Hot Encoding

### This notebook goes through how to implement one hot encoding to deal with categorical data  

```python
# import packages
import numpy as np
import pandas as pd
import seaborn as sn
```

```python
# read in the data set
data = pd.read_csv("data1.csv")
data
```

### We can see that marital status is given as Married (M), Single (S) and Divorced (D)

```python
# count plot
sn.countplot(x="Marital_Status", data=data)
```

### If this was a classification problem we need to apply a model such as logistic regression. However, we need the marital status to be numerical data so the model can take it into account. We can map the categories to numerical values. This is called one hot encoding and expands the Marital Status column to three other columns and assigns binary values.for where 1 = True and 0 = False depending on the example. The code below will explain and visualise this better.

```python
# create the dummies
dum = pd.get_dummies(data["Marital_Status"])
dum
```

```python
# combine the data
data = pd.concat([data, dum], axis="columns")
data
```

```python
# delete the unecessary column
del data["Marital_Status"]
data
```

### And there we are! We have just one encoded the data so it can be used in classification models.
