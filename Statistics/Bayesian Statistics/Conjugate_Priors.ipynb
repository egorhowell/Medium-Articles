```python
# Import packages
from scipy.stats import beta as beta_dist
import matplotlib.pyplot as plt
import numpy as np
import scipy.special as sc
```

```python
# Plot the distribution
alpha = 49
beta = 151
x = np.arange(0, 1, 0.001)
y = beta_dist.pdf(x, alpha, beta)
plt.figure(figsize=(12, 6))
plt.plot(x, y, linewidth=3)
plt.xlabel("Batting Average", fontsize=20)
plt.ylabel("PDF", fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.axvline(0.244, linestyle="dashed", color="black", label="Average")
plt.legend(fontsize=18)
plt.savefig("beta_dist.png")
plt.show()
```

```python
# Plot the distribution
alpha = 49
beta = 151
new_alpha = 49 + 3
new_beta = 151 + 0
x = np.arange(0, 1, 0.001)
y = beta_dist.pdf(x, alpha, beta)
y_new = beta_dist.pdf(x, new_alpha, new_beta)
plt.figure(figsize=(12, 6))
plt.plot(x, y, linewidth=3, label="Prior")
plt.plot(x, y_new, linewidth=3, label="Posterior")
plt.xlabel("Batting Average", fontsize=20)
plt.ylabel("PDF", fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.axvline(0.244, linestyle="dashed", color="black", label="Average")
plt.legend(fontsize=18)
plt.savefig("beta_dist.png")
plt.show()
print(sum(y_new))
```

```python
alpha = 49
beta = 151
x = np.arange(0, 1, 0.001)
prior = beta_dist.pdf(x, alpha, beta)
likelihood = 10272278170 * x**40 * (1 - x) ** 10
posterior = prior * likelihood
posterior = posterior / sum(posterior)
plt.figure(figsize=(12, 6))
plt.plot(x, posterior, linewidth=3, label="Posterior")
plt.xlabel("Batting Average", fontsize=20)
plt.ylabel("Probability", fontsize=20)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.axvline(0.244, linestyle="dashed", color="black", label="Average")
plt.legend(fontsize=18)
plt.savefig("beta_dist.png")
plt.show()
```
